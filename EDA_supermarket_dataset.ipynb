```python
sales['Branch'].value_counts().plot(kind = 'pie', autopct = '%1.2f%%', shadow = True)
plt.show()
```

```python
sales['Payment'].value_counts().plot(kind ='bar')
plt.title('Most Popular Payment Method Used by Customer')
plt.xlabel('Payment Methods')
plt.show()
```

```python
sales['Rating'].value_counts()
```

```python
sns.distplot(sales['Rating'])
plt.title('Distribution of Customer Rating')
plt.show()
```

```python
sales['Rating'].skew
```

```python
sns.distplot(sales['cogs'])
plt.show()
```

```python
sns.scatterplot(x = 'cogs', y = 'Rating',data = sales)
plt.title('Relationship between COGS and Customer Rating')
plt.xlabel('Cost of Goods Sold')
plt.ylabel('Customer Rating')
plt.show()
```

```python
sns.scatterplot(x = 'gross income', y = 'Rating', data = sales)
plt.title('Relationship between Gross Income and Customer Rating')
plt.xlabel('Gross Income')
plt.ylabel('Customer Rating')
plt.show()
```

```python
sns.barplot(x = 'Branch', y = 'gross income', data = sales, hue = 'City')
plt.title('Most Profitable Branch per Gross Profit')
plt.xlabel('Branch')
plt.ylabel('gross income', fontsize = 12)
plt.show()
```

```python
sns.barplot(x = 'Product line', y = 'gross income', data = sales)
plt.xticks(rotation = 80)
plt.title('Product line with the most Gross Income')
plt.xlabel('Product Line')
plt.ylabel('Gross Income')
plt.show()
```

```python
sns.barplot(x = 'Product line', y = 'Unit price', data = sales)
plt.xticks(rotation = 80)
plt.title('Product line with the Highest unit price')
plt.xlabel('Product Line')
plt.ylabel('Unit Price')
plt.show()
```

```python
pd.crosstab(sales['City'], sales['Payment'])
```

```python
sns.heatmap(pd.crosstab(sales['City'], sales['Payment']))
```

```python
sales.groupby('Product line').sum()['Quantity'].plot(kind = 'bar')
plt.title('Product line with the Highest Quantity')
plt.xlabel('Product Line')
plt.ylabel('Quantity')
plt.show()
```
